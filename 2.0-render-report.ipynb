```python
import json
import pandas as pd
from IPython.core.display import HTML

pd.set_option("display.max_columns", 500)
pd.set_option("display.max_colwidth", 500)
pd.options.display.max_rows = 999
```

```python
with open('data.txt','r') as f:
    data = json.load(f)
```

```python
df = pd.DataFrame(data)
```

```python
df = pd.DataFrame(data).drop(['branding', 'agents', 'office', 'client_display_flags', 'lead_forms', 'list_tracking'], axis='columns')
```

```python
cols = df.columns.tolist()
```

```python
cols = [cols[cols.index("thumbnail")]] + cols[:cols.index("thumbnail")] + cols[cols.index("thumbnail")+1:]
```

```python
df = df[cols]
```

```python
df['sqft'] = df['building_size'].apply(pd.Series)['size']
```

```python
df['sqft_cost'] = df.price / df.sqft
df['bed_cost'] = df.price/df.beds
```

```python
# Converting links to html tags
def path_to_image_html(path):
    return '<img src="'+ path + '" width="500" >'

def path_to_anchor_html(path):
    return '<a href="'+ path + '" >Link to Posting</a>'
```

```python
df = df.sort_values('sqft_cost')
```

```python
#df = df[df['prop_type'] == 'multi_family']
```

```python
HTML(df.to_html(escape=False,formatters=dict(thumbnail=path_to_image_html, rdc_web_url=path_to_anchor_html)))
```

```python
df.to_html('results.html',escape=False, formatters=dict(thumbnail=path_to_image_html, rdc_web_url=path_to_anchor_html))
```
